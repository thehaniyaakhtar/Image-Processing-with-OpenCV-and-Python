```python
import pandas as pd
import numpy as np

import glob 

import matplotlib.pyplot as plt

import cv2 
```

```python
dog_files = glob.glob("../input/dogs-cats-images/dataset/training_set/dogs/*.jpg")
cat_files = glob.glob("../input/dogs-cats-images/dataset/training_set/cats/*.jpg") 
```

```python
#reading an image file
img_mpl = plt.imread(cat_files[20])
img_cv2 = cv2.imread(cat_files[20])
img_mpl.shape, img_cv2.shape
```

```python
#h, w, channels

#usually about 255
img_mpl.max()

#dividing the pixels with 255 to get a normal number between 0 and 1
```

```python
#flattening the values into one long array
img_mpl.flatten()

#making it a pandas series
pd.Series(img_mpl.flatten())

#plotting it on a Histogram
pd.Series(img_mpl.flatten()).plot(kind='hist', bins=50, title='Dist of pixel values')
```

```python
#displaying images
fig, ax = plt.subplots(figsize = (10, 10))  
ax.imshow(img_mpl)
ax.axis('off')#turning the grid off
plt.show()
```

```python
#image channels
#stacks of r, g, b colored numpy arrays
#(h, w, 3 channels)
img_mpl.shape
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img_mpl[:, :, 0], cmap='Reds')
axs[0].set_title('Red Channel')
axs[1].imshow(img_mpl[:, :, 1], cmap='Greens')
axs[0].set_title('Green Channel')
axs[2].imshow(img_mpl[:, :, 2], cmap='Blues')
axs[0].set_title('Blue Channel')
```

```python
#images look diff w cv2 and matplotlib as they both load channels din diff orders
#cvs: b, g, r
#mlp: r, g, b
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 5))
axs[0].imshow(img_cv2)
axs[0].axis('off')
axs[1].axis('off')
axs[1].imshow(img_mpl)
axs[0].set_title("CSV Image")
axs[1].set_title("MLP Image")
plt.show()
```

```python
#Converting bgr to rgb
img_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots()
ax.imshow(img_cv2_rgb)
ax.axis('off')
plt.show()
```

```python
#image manipulation
img = plt.imread(cat_files[7])
fig, ax = plt.subplots(figsize = (5, 8))
ax.imshow(img)
ax.axis('off')
plt.show()
```

```python
#RGB@GRAY
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(img_gray, cmap = 'Greys')
ax.axis('off')
ax.set_title('Grey Image')
plt.show()
```

```python
#Resizing and Scaling
img_resized = cv2.resize(img, None, fx = 0.25, fy = 0.25)
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(img_resized)
ax.axis('off')
ax.set_title('Image Resize')
plt.show()
```

```python
img_resized.shape
```

```python
#giving new height and width instead of noen
img_resize = cv2.resize(img, (100, 200))
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(img_resize)
ax.axis('off')
ax.set_title('Image Resize')
plt.show()
```

```python
#interpolation
cv2.resize(img, (5000, 5000), interpolation = cv2.INTER_CUBIC)
```

```python
img_resize.shape, img.shape
```

```python
#applying different kernels to images will produce diff results
#filtering, blurring sharpening

#sharpen image
kernel_sharpening = np.array([[-1, -1, -1],
                              [-1, 9, -1],
                              [-1, -1, -1]])
sharpened = cv2.filter2D(img, -1, kernel_sharpening)
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(sharpened)
ax.axis('off')
ax.set_title('Image Sharpened')
plt.show()
```

```python
#Blurring the image
kernels = np.ones((3, 3), np.float32)/9
kernels
blurred = cv2.filter2D(img, -1, kernels)
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(blurred)
ax.axis('off')
ax.set_title('Image Blurred')
plt.show()
```

```python
#Saving image
plt.imsave('mpl_dog.png', blurred)
cv2.imwrite('cv2_dog.png', blurred) #provides output
```
